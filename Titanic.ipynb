```python
import csv
with open("/content/drive/MyDrive/titanic.csv",'r') as arquivo:
    leitor = csv.reader(arquivo)
    dados = list(leitor)

# Sobreviventes por sexo
def sobreviventesPorSexo(dados):  
  countHomens = 0    
  countMulheres = 0    
  for x in range(1,len(dados)):
    if (dados[x][1] == '1' and dados[x][4] == 'male'):
      countHomens += 1
    elif(dados[x][1] == '1' and dados[x][4] == 'female'):
      countMulheres += 1

  return print(f'Sobreviventes:\nHomens: {countHomens}\nMulheres: {countMulheres}')

# Idade média dos passageiros
def mediaIdade(dados):
  total = 0
  for x in range(1,len(dados)):
    if(dados[x][5] != 'NaN' and dados[x][5]!= ''):
      total = total + (float(dados[x][5]))
  idadeMedia = total/len(dados)
  return print(f'Idade média dos passageiros: {idadeMedia:,.2f} Anos')

# Média dos valores das passagens
def mediaPassagem(dados):
  total = 0
  for x in range(1,len(dados)):
      total = total + (float(dados[x][9]))
  mediaValor = total/len(dados)
  return print(f'Valor médio das passagens: $ {mediaValor:,.2f}')

# Sobreviventes
def sobreviventes(dados):
  countSobreviventes = 0    
  for x in range(1,len(dados)):
    if (dados[x][1] == '1'):
      countSobreviventes += 1
  return print(f'Número de Sobreviventes: {countSobreviventes}')

#Contador de Mães
def maes(dados):
  totalMaes = 0
  for x in range(1, len(dados)):
    if(dados[x][7] != '0' and dados[x][4] == 'female'):
        totalMaes += 1
  return print(f'Mães: {totalMaes}\n')

#Contador de Órfãos
def orfaosPaisMaes(dados):
  perdeuMae = 0
  perdeuPai = 0
  for x in range(1, len(dados)):
    if(dados[x][7] != '0' and dados[x][4] == 'female'):
        perdeuMae = perdeuMae + (float(dados[x][7]))
    elif(dados[x][7] != '0' and dados[x][4] == 'male'):
        perdeuPai = perdeuPai + (float(dados[x][7]))
  return print(f'Filhos que perderam a mãe: {perdeuMae}\nFilhos que perderam o pai: {perdeuPai}\n')

#Contador de Crianças
def criancas(dados):
  totalCriancas = 0
  idadeTotal = 0
  for x in range(1, len(dados)):
    if(dados[x][1] == '0' and dados[x][5] <= '11'):
        totalCriancas += 1
  return print(f'Total de crianças: {totalCriancas}\n')

#Taxa de sobrevivência por sexo
def taxaSobrevivenciaSexo(dados):
  taxaMasculino = 0
  totalMasculino = 0
  taxaFeminino = 0
  totalFeminino = 0;
  for x in range(1, len(dados)):
    if(dados[x][4] == 'male'):
      totalMasculino += 1
    if(dados[x][4] == 'female'):
      totalFeminino += 1
    if(dados[x][1] == '1' and dados[x][4] == 'male'):
        taxaMasculino += 1
    if(dados[x][1] == '1' and dados[x][4] == 'female'):
        taxaFeminino += 1
  return print(f'Taxa de sobrevivência:\nMasculina: {(taxaMasculino/totalMasculino)*100:,.2f}%\nFeminina: {(taxaFeminino/totalFeminino)*100:,.2f}% \n')

#Quantidade de passageiros por classe
def quantidadeClasse(dados):
  totalClasseA = 0
  totalClasseB = 0
  totalClasseC = 0
  for x in range(1, len(dados)):
    if(dados[x][2] == '1'):
      totalClasseA += 1
    elif(dados[x][2] == '2'):
      totalClasseB += 1
    elif(dados[x][2] == '3'):
      totalClasseC += 1
  return print(f'Quantidade de passageiros por classe:\nClasse A: {(totalClasseA/len(dados))*100:,.2f}%\nClasse B: {(totalClasseB/len(dados))*100:,.2f}% \nClasse C: {(totalClasseC/len(dados))*100:,.2f}% \n')

#Taxa de sobrevivência por classe
def taxaSobrevivenciaClasse(dados):
  taxaClasseA = 0
  taxaClasseB = 0
  taxaClasseC = 0
  totalClasseA = 0
  totalClasseB = 0
  totalClasseC = 0;
  for x in range(1, len(dados)):
    if(dados[x][2] == '1'):
      totalClasseA += 1
    elif(dados[x][2] == '2'):
      totalClasseB += 1
    elif(dados[x][2] == '3'):
      totalClasseC += 1
    if(dados[x][1] == '1' and dados[x][2] == '1'):
       taxaClasseA += 1
    elif(dados[x][1] == '1' and dados[x][2] == '2'):
       taxaClasseB += 1
    elif(dados[x][1] == '1' and dados[x][2] == '2'):
       taxaClasseC += 1
  return print(f'Taxa de sobrevivência:\nClasse A: {(taxaClasseA/totalClasseA)*100:,.2f}%\nClasse B: {(taxaClasseB/totalClasseB)*100:,.2f}% \nClasse C: {(taxaClasseC/totalClasseC)*100:,.2f}% \n')

print("## Menu Titanic ##")
print("1) Sobreviventes por sexo! ")
print("2) Média de idade dos passageiros! ")
print("3) Média de valor das passagens ")
print("4) Número de sobreviventes! ")
print("5) Mães abordo! ")
print("6) Total de crianças! ")
print("7) Quantidade de órfãos! ")
print("8) Quantidade por classe! ")
print("9) Taxa de sobrevivência por sexo!")
print("10) Taxa de sobrevivência por classe!")

opcao = str(input("Digite sua opção:"))
if opcao == '1':
    a = sobreviventesPorSexo(dados)
if opcao == '2':
    b = mediaIdade(dados)
if opcao == '3':
    c = mediaPassagem(dados)
if opcao == '4':
    d = sobreviventes(dados)
if opcao == '5':
    e = maes(dados)
if opcao == '6':
    f = criancas(dados)
if opcao == '7':
    g = orfaosPaisMaes(dados)
if opcao == '8':
    h = quantidadeClasse(dados)
if opcao == '9':
    i = taxaSobrevivenciaSexo(dados)
if opcao =='10':
    j = taxaSobrevivenciaClasse(dados)
```
